```python
import dash
from dash import Dash,html,dcc,callback
from dash.dependencies import Input,Output,State
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
app =Dash()

app.layout = html.Div(children=[
    html.H1('Welcome', style={'text-align': 'center', 'color': 'blue'}),
    html.H3('Please choose your Borough'),
    dcc.Dropdown(id='in1',options=df.borough.unique(),value=''),

    dcc.Graph(id='out1', style={'width': '48%', 'display': 'inline-block'}),
    dcc.Graph(id='out2', style={'width': '48%', 'display': 'inline-block'}),
    dcc.Graph(id='out3', style={'width': '48%', 'display': 'inline-block'}),
    dcc.Graph(id='out4', style={'width': '48%', 'display': 'inline-block'}),
    dcc.Graph(id='out5', style={'width': '48%', 'display': 'inline-block'}),
    dcc.Graph(id='out6', style={'width': '48%', 'display': 'inline-block'}),
    dcc.Graph(id='out7', style={'width': '48%', 'display': 'inline-block'}),
    dcc.Graph(id='out8', style={'width': '48%', 'display': 'inline-block'})
])

@callback(
    [Output('out1', 'figure'),
     Output('out2', 'figure'),
     Output('out3', 'figure'),
     Output('out4', 'figure'),
     Output('out5', 'figure'),
     Output('out6', 'figure'),
     Output('out7', 'figure'),
     Output('out8', 'figure')],
     Input('in1', 'value')

)

def update_graph(value1):
    filtered_df = df[df['borough'] == value1]


    fig1 = px.bar(filtered_df, x='borough', y='number of persons injured',           
                title='Number of Persons Injured by Borough',
                color='number of persons injured', color_continuous_scale='RdYlBu')

    fig2 = px.bar(filtered_df, x='borough', y='number of persons killed',
                title='Number of Persons Killed by Borough',
                color='number of persons killed', color_continuous_scale='RdYlBu')

    fig3 = px.bar(filtered_df, x='borough', y='number of pedestrians injured',
                title='Number of Pedestrians Injured by Borough',
                color='number of pedestrians injured', color_continuous_scale='RdYlBu')

    fig4 = px.bar(filtered_df, x='borough', y='number of pedestrians killed',
                title='Number of Pedestrians killed by Borough',
                color='number of pedestrians killed', color_continuous_scale='RdYlBu')

    fig5 = px.bar(filtered_df, x='borough', y='number of cyclist injured',
                title='Number of Cyclist Injured by Borough',
                color='number of cyclist injured', color_continuous_scale='RdYlBu')

    fig6 = px.bar(filtered_df, x='borough', y='number of cyclist killed',
                title='Number of Cyclist killed by Borough',
                color='number of cyclist killed', color_continuous_scale='RdYlBu')

    fig7 = px.bar(filtered_df, x='borough', y='number of motorist injured',
                title='Number of Motorist Injured by Borough',
                color='number of motorist injured', color_continuous_scale='RdYlBu')

    fig8 = px.bar(filtered_df, x='borough', y='number of motorist killed',
                title='Number of Motorist Injured by Borough',
                color='number of motorist killed', color_continuous_scale='RdYlBu')
    
    return fig1, fig2, fig3, fig4, fig5, fig6, fig7, fig8
    
app.run(jupyter_mode='external')
```
